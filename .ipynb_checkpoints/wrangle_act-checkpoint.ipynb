### QUALITY

### Issue #1: Rating_Denominator and Rating_Numerator [df_twitter_archive] Columns: Inaccurate rating values

#### Define: 

The tweets' text have the accurate ratings embedded in them.

Using regex, extract the accurate rating values from the tweets' text and set the accurate values to be the `rating_numerator` and `rating_denominator`

#### Code

```python
## EXTRACT ACCURATE RATINGS FROM TWEETS

# empty list to hold all ratings
ratings = []
# pattern to extract ratings
rating_pattern = r'(\d+\.?[\d]*/\d+\.?[\d]*[/\d]*)'
# loop through the text column of the dataframe
for text in archive_copy['text']:
    # get all instances of rating-like pattern from tweets
    rating_match = re.findall(rating_pattern, text) 
    # There is 1 tweet where there are 3 ratings;
    #'10/10 for dog. 7/10 for cat. 12/10 for human. Much skill. Would pet all https://t.co/uhx5gfpx5k'
    # The first rating is picked.
    if len(rating_match) > 2:
         ratings.append(rating_match[0])
    # if there are more than 1 rating-like matches, 
    # the 2nd one is often the rating for the dog
    elif len(rating_match) > 1:
        ratings.append(rating_match[-1])
    # if there is only one match
    elif len(rating_match) == 1:
        # if match is like any of the none rating values
        if rating_match[0] in ('24/7', '9/11', '7/11') or\
        len(re.findall(r'/',rating_match[0])) > 1: # or rating match contains more than 1 '/' - which may be a date
            # set rating for this row to be 0
            ratings.append('0/0')
        else:
            # otherwise, add rating to ratings
            ratings.append(rating_match[0])
    # no match found
    else:
        # set rating for this row to be 0
        ratings.append('0/0')
```

#### Test

```python
ratings[:20]
```

```python
# create new column called ratings and assign 
# ratings list to it.
twitter_archive['ratings'] = ratings
```

#### Remove links from `text`

```python
def remove_links(text):
    """
    Function to remove links from text
    Takes in a text that possibly contains links as input
    Uses regex to find links
    Returns a text devoid of url links
    """
    # create link pattern
    link_pattern = r'https://[a-zA-Z0-9/\.]+'
    # substitute link in text with blank
    # up to 4 links
    text = re.sub(link_pattern, '', text, count=4)
    
    # return text without links
    return text

# apply remove links
df_subset['text'] = df_subset['text'].apply(remove_links)
```

#### Remove ratings from `text`

```python
def remove_ratings(text):
    """
    Function to remove ratings from text
    Takes in a text containing rating in form '10/10'
    Uses regex to find ratings
    Returns a text devoid of ratings
    """
    # substitute rating in text with blank
    # up to 4 ratings
    text = re.sub(rating_pattern, '', text, count=4)
    
    # return text without ratings
    return text

# apply remove ratings
df_subset['text'] = df_subset['text'].apply(remove_ratings)
```

#### Remove stopwords from `text` column

```python
def remove_stopwords(text):
    """
    Function to remove stopwords from text
    Takes in a text containing stopwords as input
    Removes stopwords
    Returns a text without stopwords
    """
    # initialize an empty list to hold words
    # that are not stopwords
    new_text = []
    # iterate through the list of words
    # in the text
    for word in text.split(' '):
        # checks if a word is not in the stopwords list
        if word not in set(STOPWORDS):
            new_text.append(word) # appends the word to the empty list
    
    # returns a text without stopwords
    return " ".join(new_text)

# apply remove stopwords
df_subset['text'] = df_subset['text'].apply(remove_stopwords)
```

#### Remove Punctuations from `text` column

```python
def remove_punctuations(text):
    """
    Function to remove punctuations from text
    Takes in a text containing punctuations as input
    Removes punctuations
    Returns a text without punctuations
    """
    # iterate through each character in the text
    for char in text:
        # checks if the charecter is a punctuation
        if char in list(punctuation):
            text = text.replace(char, '') # replaces punctuation with blank
    
    # returns text without punctuations
    return text

# apply remove punctuations
df_subset['text'] = df_subset['text'].apply(remove_punctuations)
```

```python
def aggregate_words(df, column):
    """
    Function to aggregate words in a column that has list values
    Takes in a dataframe and a column as inputs
    Returns a flattened list containing all the words in the column
    """
    # initialize an empty list to hold words
    all_words = []
    # iterate through the column of the dataframe
    for arr in df[column]:
        # check if column iter is a list
        if isinstance(arr, list):
            # iterate through column iter
            for word in arr:
                # append values to empty list
                all_words.append(word)
        # if column iter is not a list
        else:
            # append value to empty list
            all_words.append(arr)
    
    # return flattened list
    return all_words

def make_freq_tbl(arr):
    """
    Function to create a frequency dictionary
    Takes in a list as input
    Returns a frequency dictionary
    """
    # create empty dictionary to hold the 
    # frequency distribution
    freq_tbl = {}
    # iterate through the list
    for word in arr:
        # checks if the arr iter is in the freq_tbl
        if word in freq_tbl:
            # adds 1 to frequency value of the iter
            freq_tbl[word] += 1
        # if the arr iter is not in the freq_tbl
        else:
            # sets the frequency value of the iter to 1
            freq_tbl[word] = 1
    
    # return frequency dictionary
    return freq_tbl


def word_cloud(freq_tbl):
    """
    Function to create a word cloud visualization
    Takes in a frequency dictionary as input
    Returns a word cloud visualization
    """
    # create the word cloud object
    cloud = WordCloud(width = 600, height = 400, # set image width and height
                    background_color ='black', # set background color
                    stopwords = set(STOPWORDS), # set stopwords package
                    min_font_size = 10) # set minimum font size of words
    # generate wprd cloud
    cloud.generate_from_frequencies(freq_tbl)

    # plot the WordCloud image                      
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(cloud)
    plt.axis("off")
    plt.tight_layout(pad = 0.2)
    plt.savefig('visuals_for_project_michael/word_cloud_' + str(time.time()) + '.png')
    plt.show()
```

```python
# make words soup
all_words = aggregate_words(top_rated_dogs, 'text')
# create frequency table
freq_tbl_best = make_freq_tbl(all_words)
# generate word cloud
word_cloud(freq_tbl_best)
```

##### Top 10 words that are associated with a high rating

```python
def get_top_n(obj, n=10):
    """
    Function to get top n values of a frequency dictionary
    :params
    - obj: frequency dictionary
    - n: number of top values to output
    Returns a dictionary of top n values
    """
    # sort dictionary value in descending manner
    sorted_obj = {k: v for k, v in sorted(obj.items(), reverse=True, key=lambda x: x[1])}
    # get first n values
    top_n = dict(list(sorted_obj.items())[:n])
    
    # return first n values   
    return top_n

def dict2df(obj):
    """
    Function to convert dictionary to a dataframe
    Takes in a frequency dictionary as input
    Returns a dataframe
    """
    # create a dataframe from a dictionary
    df = pd.DataFrame.from_dict(obj, orient='index')
    # return dataframe
    return df

def make_barplot(df, xlabel='Words'):
    """
    Function to create a barplot visualization
    Takes in a dataframe and user-defined xlabel of the plot as input
    Returns a barplot visualization
    """
    # plot the barplot
    plot = df.plot(kind='bar', color='Blue', figsize=(12, 6))
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel('Frequency', fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.title(f"Frequency of {xlabel.title()}", fontsize=18)
    plt.tight_layout(pad = 0.2)
    plt.savefig('visuals_for_project_michael/' + xlabel + 'barplot_' + str(time.time()) + '.png')
    # return plot
    return plot
    
```

```python
# get top 10 values
top_10 = get_top_n(freq_tbl_best)
# convert top_10 dictionary to dataframe
top_10_df = dict2df(top_10)
# plot barplot of dataframe
make_barplot(top_10_df);
```

```python
def make_scatterplot(df, x, y):
    """
    Function to create a scaterplot visualization
    Takes in a dataframe and the 2 columns to correlate as inputs
    Returns a scatterplot visualization
    """
    # create scatterplot
    plt.figure(figsize=(12,6))
    plot = sns.regplot(data=df, x=x, y=y, color='Blue') # regplot is a scatterplot with a regression line
    plt.xlabel(x, fontsize=14)
    plt.ylabel(y, fontsize=14)
    plt.title(f"Relationship Between {x.replace('_', ' ').title()} and {y.replace('_', ' ').title()}", fontsize=18)
    plt.tight_layout(pad = 0.2)
    plt.savefig(fname='visuals_for_project_michael/' + x + '_' + y + '_scatterplot_' + str(time.time()) + '.png')
    # return plot
    return plot
```

```python
# plot scatterplot
make_scatterplot(df_subset, 'rating_numerator', 'retweet_count');
```

```python
def get_corrcoef(df, x, y):
    """
    Takes two columns to be correlated against each other
    Return a correlation coefficient matrix as a heatmap
    """
    # get popularity column
    x_series = df[x]
    # get efficiency column
    y_series = df[y]
    # get correlation coefficient of popularity and efficiency
    correlation = np.corrcoef(x_series, y_series)
    #plot correlation matrix
    plt.figure(figsize=(12, 6))
    plot = sns.heatmap(correlation, cmap="Blues", annot=True, 
                       cbar=False, xticklabels=[x, y],
                      yticklabels=[x, y], annot_kws={'fontsize': 16})
    plt.title(f"Correlation Between {x.replace('_', ' ').title()} and {y.replace('_', ' ').title()}", fontsize=18)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout(pad = 0.2)
    plt.savefig(fname='visuals_for_project_michael/' + x + '_' + y + '_corr_' + str(time.time()) + '.png')
    # return correlation coefficient plot
    plt.show()
```

```python
# plot correlation heatmap
get_corrcoef(df_subset, 'rating_numerator', 'retweet_count');
```

There is relatively no correlation (__0.052__) between the rating given to a dog photo and the number of retweets the photo gets. The ratings of a dog photo has no influence on the number of retweets it gets. This can very well be accredited to the outliers in the rating numerators.

### Predicting Ratings

#### Stemming

```python
# create the stemmer object with the required
# language parameter set
stemmer = SnowballStemmer(language='english')
```

```python
def stem(arr):
    """
    Function to stem words in an array
    Takes in a list as input
    Returns a stemmed array of words
    """
    # initialize an empty list to hold
    # stemmed words
    stemmed = []
    # iterate through array
    for word in arr:
        # stem each word in the array
        stemmed_word = stemmer.stem(word)
        # append stemmed word to empty list
        stemmed.append(stemmed_word)
        
    # return stemmed array
    return stemmed
```

```python
# apply stemming function
df_subset['text'] = df_subset['text'].apply(stem)
```

#### Train-Test Split

```python
# set data needed for modelling 
# into the X and y variables
X = df_subset['text']
y = df_subset['rating_numerator'].astype(int)
```

```python
# split data into training sets and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

#### Vectorization

```python
def make_corpus(arr):
    """
    Function to convert list to sring of words
    Takes in a list as input
    Returns a string of words
    """
    # return string of words
    return " ".join(arr)
```

```python
# apply the make corpus function
X_train_corpus = X_train.apply(make_corpus)
X_test_corpus = X_test.apply(make_corpus)
# create the vectorizer object
vectorizer = CountVectorizer()
# Fit and transform the training sets
X_train = vectorizer.fit_transform(X_train_corpus)
X_test = vectorizer.transform(X_test_corpus)
```

#### Modelling

```python
# creating a RF classifier
clf = RandomForestClassifier(n_estimators = 100, random_state=42)
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)
```

```python
# performing predictions on the train dataset
train_pred = clf.predict(X_train)
# performing predictions on the test dataset
test_pred = clf.predict(X_test)
```

#### Model Accuracy

```python
# train accuracy
train_accuracy = accuracy_score(train_pred, y_train)
# test accuracy
test_accuracy = accuracy_score(test_pred, y_test)
```

```python
print(f'The train accuracy is: {round(train_accuracy * 100, 2)}%')
print()
print(f'The test accuracy is: {round(test_accuracy * 100, 2)}%')
```
